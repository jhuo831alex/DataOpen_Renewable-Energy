```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('energy_consumption.csv')
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df.head()
```

```python
df_2 = pd.read_csv('state_area_final.csv')

#df_2['state_name'].unique()
df_3=df.merge(df_2, on = 'state_name')
df_3.head()
```

```python
df_3 = df_3.drop(columns ='Unnamed: 0')
```

```python
df_3.head()
```

```python
df_3['Renewable Consumption per Area'] = np.array(df_3['Renewable Consumption(BTU)']) / np.array(df_3['area_data'])
df_3['Non-renewable Consumption per Area'] = np.array(df_3['Non-renewable Consumption(BTU)']) / np.array(df_3['area_data'])

df_3.head()
```

```python
df_3.to_csv('consumption_per_area.csv')
```
