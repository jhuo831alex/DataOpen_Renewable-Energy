```python
import pandas as pd
from pandas import DataFrame, Series
```

```python
year = range(2000, 2019)
```

```python
df = pd.read_csv("annual_aqi_by_county_1999.csv")
for y in year:
    df2 = pd.read_csv("annual_aqi_by_county_" + str(y) + ".csv")
    df = pd.concat([df, df2])
```

```python
s = df.groupby(["State", "Year"])["Max AQI", "Median AQI"].mean()
```

```python
s.to_csv("annual_aqi_by_state.csv")
```
