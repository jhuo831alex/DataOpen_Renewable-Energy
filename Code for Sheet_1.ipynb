```python
idx = (df['energy_bin']=='Renewable') & (df['unit'] =='Billion Btu' )
df[idx]['msn'].unique()
```

```python
# Non-Renewable energy Code:

Energy_code = df['msn']
#list non-renewable energy
#Non_Rew = ['BM', 'BQ', 'BY', 'EL', 'EM', 'EN', 'EQ', 'EY', 'GE', 'HY', 'IQ', 'IY', 'LO', 'NU', 'PL'] 

Rew = 'RE' #renewable energy code
Total = 'TE' # total
pattern_rew = r'\w{4}V$' #renewable expenditure
pattern_total = r'^TE\w{2}V$' # total energy expenditure
renew_col = []
non_new_col = []
for u in Energy_code.unique():
    if re.findall(pattern_rew, u) != [] :
        renew_col.append(u)
    else:
        continue
for u in Energy_code.unique():
    if re.findall(pattern_total, u) != [] :
        non_new_col.append(u)
    else:
        continue      
```

```python
df_renew = df[df['msn'] == 'RETCB'] # use renewable energy total consumption cost in BTU
df_renew.sort_values(['state_code','year'], inplace = True)
df_renew.info()
```

```python
df_total = df[df['msn'] == 'TETCB'] # get total energy total consumption cost in BTU
df_total.sort_values(['state_code','year'], inplace = True)
df_total.info()
```

```python
df_renew['Non-renewable Consumption(BTU)'] = np.array(df_total['value']) - np.array(df_renew['value'])
df_renew.info
```

```python
df_renew = df_renew.rename(index=str,columns={"value": "Renewable Consumption(BTU)"})
df_renew.reset_index()
```

```python
df_renew.drop(columns=['index', 'msn', 'description','energy_bin'],inplace=True)
```
